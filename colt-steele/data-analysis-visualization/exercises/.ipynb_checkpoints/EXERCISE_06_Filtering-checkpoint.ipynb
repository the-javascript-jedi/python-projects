# Filtering Data Exercise
Answer the following questions using the `bestsellers.csv` dataset

```python
import pandas as pd
books=pd.read_csv("../data/bestsellers.csv")
books
```

```python
#Find the books written by Pete Souza
pete_books=books["Author"]=="Pete Souza"
books[pete_books]
```

```python
# Find the books that are under 10 dollars
under_10_price=books["Price"]<=10
books[under_10_price]
```

```python
#Find the books that have a price between 50 and 60 dollars
```

```python
# Method 1
books[books["Price"].between(50,60)]
```

```python
# Method 2
books[(books["Price"]>50)&(books["Price"]<60)]
```

```python
#Find all the books written by Kristin Hannah, Andy Weir, or Delia Owens
```

```python
#Method 1
kristin=books["Author"]=="Kristin Hannah"
andy=books["Author"]=="Andy Weir"
delia=books["Author"]=="Delia Owens"
books[(kristin)|(andy)|(delia)]
```

```python
#Method 2
books[books["Author"].isin(["Kristin Hannah","Andy Weir","Delia Owens"])]
```

```python
#Find the Non Fiction books that are rated 4.9
books[(books["User Rating"]==4.9)&(books["Genre"]=="Non Fiction")]
```

```python
#Find the fiction book with the lowest User Rating
books[books["Genre"]=="Fiction"].sort_values("User Rating").head(1)
```

```python
#Find 2012's top 5 Fiction books with the most Reviews
df=books[(books["Year"]==2012)&(books["Genre"]=="Fiction")]
df.sort_values("Reviews",ascending=False).head(5)
```
