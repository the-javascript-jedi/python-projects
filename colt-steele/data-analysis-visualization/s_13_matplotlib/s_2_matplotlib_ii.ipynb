```python
import matplotlib.pyplot as plt
import numpy as np
nums=np.arange(5)
plt.plot(nums,nums)
```

```python
# custom title
salaries=[55000,65000,72000,90000,115000,150000]
ages=[20,25,30,32,40,45]
plt.plot(ages,salaries)
plt.title("Company Salaries")
# create a new figure and provide a distinct title
plt.figure()
plt.plot(nums,nums,color="olive",linewidth=4,linestyle="dashed")
# move title to right side
plt.title("Linear Stuff",loc="right")
```

```python
# change label of x axis
salaries=[55000,65000,72000,90000,115000,150000]
ages=[20,25,30,32,40,45]
plt.plot(ages,salaries)
plt.title("Company Salaries",color="green")
# labelpad - provides label spacing
plt.xlabel("Age",labelpad=10)
plt.ylabel("Average Salary",labelpad=10,fontsize=24,rotation="vertical")
```

# Changing X & Y Ticks

```python
# change label of x axis
salaries=[55000,65000,72000,90000,115000,150000]
ages=[20,25,30,32,40,45]
plt.plot(ages,salaries)
plt.title("Company Salaries",color="green")
# manual ticks for x and y axis
plt.xticks([20,25,30,35,40,45])
plt.yticks([60000,80000,100000,120000,140000],labels=["60k","80k","100k","120k","140k"])
```

# Adding Legends To Plots

```python
plt.figure(figsize=(6,6))
plt.plot(nums,color="teal", label="x")
plt.plot(nums**2,color="olive", label="x squared")
plt.plot(nums**3,color="purple", label="x cubed")
plt.legend()
```

```python
# legend styles
plt.figure(figsize=(6,6))
plt.plot(nums,color="teal", label="x")
plt.plot(nums**2,color="olive", label="x squared")
plt.plot(nums**3,color="purple", label="x cubed")
plt.legend(loc="center left",shadow=True,frameon=True,facecolor="white")
```
