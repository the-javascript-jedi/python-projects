```python
import matplotlib.pyplot as plt
import numpy as np
nums=np.arange(5)
```

```python
plants = ['Spinach', 'Turnip', 'Rhubarb', 'Broccoli', 'Kale']
died = [10,25,5,30,21]
germinated = [74, 88, 56,69,59]
```

```python
plt.figure(figsize=(6,6))
# specify width of the barchart
plt.bar(plants, germinated, color="orchid",width=0.2)
plt.title("Germination Rates")
plt.ylabel("Number Of Seeldings")
```

```python
plt.figure(figsize=(6,6))
plt.bar(plants, died, color="#3498db", bottom=germinated )
plt.bar(plants, germinated, color="#2ecc71")
```

```python
# Use this data to generate the plot!
resorts = ["Crested Butte", "Aspen", "Telluride", "Wolf Creek"]
snow_last_24 = [29,15,15,23]
snow_next_24 = [16,6,19,30]
```

```python
plt.figure(figsize=(8,6),dpi=200)
plt.bar(resorts,snow_last_24, color="#74b9ff",label="last 24")
plt.bar(resorts,snow_next_24, color="#7ed6df",bottom=snow_last_24,label="next 24")
plt.ylabel("Inches",labelpad=10)
plt.title("CO Snow Storm Report")
plt.legend()
```
