```python
import matplotlib.pyplot as plt
```

```python
plt.plot()
```

```python
import numpy as np
# numpy.arrange will provide a list of numbers evenly spaced from 0 to n
nums=np.arange(5)
```

```python
plt.plot(nums,nums)
```

```python
plt.plot(nums,nums)
plt.plot(nums,nums*nums)
```

```python
plt.plot(nums,nums)
plt.plot(nums,nums*nums)
plt.plot(nums,nums*nums*nums)
```

```python
# everything that comes after the plt.figure will be cast in a new figure
```

```python
plt.figure()
plt.plot(nums,nums)
plt.figure()
plt.plot(nums,nums*nums)
plt.figure()
plt.plot(nums,nums*nums*nums)
```

```python
plt.figure(figsize=(2,6))
plt.plot(nums,nums)
```

```python
plt.figure(figsize=(20,6))
plt.plot(nums,nums)
```

```python
plt.figure(figsize=(100,6))
plt.plot(nums,nums)
```

```python
# using dpi we can adjust the zoom
# default dpi is 80
plt.figure(figsize=(2,6),dpi=200)
plt.plot(nums,nums)
```

```python
# using dpi we can adjust the zoom
# if we use dpi 20, the plot image will be very small
plt.figure(figsize=(2,6),dpi=20)
plt.plot(nums,nums)
```

```python
# make a square size
plt.figure(figsize=(6,6))
plt.plot(nums,nums)
```

```python
plt.plot(nums,nums)
plt.plot(nums,nums*nums)
plt.plot(nums,nums*nums*nums)
```

```python
# providing a custom color
plt.plot(nums,nums,color="olive")
plt.plot(nums,nums*nums,color="#ff6b6b")
plt.plot(nums,nums*nums*nums,c="#ff9f43")
```

```python
# provide a custom line width
plt.plot(nums,nums,color="olive")
plt.plot(nums,nums*nums,color="#ff6b6b")
plt.plot(nums,nums*nums*nums,c="#ff9f43",linewidth=4)
```

```python
# provide a custom line style
plt.plot(nums,nums,color="olive",linewidth=4,linestyle="dashed")
plt.plot(nums,nums*nums,color="#ff6b6b",linewidth=4,linestyle="dotted")
plt.plot(nums,nums*nums*nums,c="#ff9f43",linewidth=4,linestyle="-.") # -. - dash -dotted line
```

```python
# provide custom marker
plt.plot(nums,nums,color="olive",marker="*",markersize=20,markerfacecolor="#ff9f43")
```
