```python
import pandas as pd
titanic=pd.read_csv("../data/titanic.csv")
titanic
```

```python
titanic.info()
```

```python
# age has both number and ? so the non number is displayed as ? (i.e object)
titanic["age"].value_counts()
```

```python
# replace the ? as None type
titanic["age"].replace(["?"],[None],inplace=True)
```

```python
titanic.age.value_counts(dropna=False)
```

```python
# create a new column age_float and convert the age
titanic["age_float"]=titanic["age"].astype("float")
titanic
```

```python
# modify the age column itself
titanic["age"]=titanic["age"].astype("float")
```
