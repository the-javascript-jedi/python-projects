```python
houses=pd.read_csv("../data/kc_house_data.csv")
houses.head()
```

```python
# houses that are waterfront but also cheap
```

```python
houses[houses["waterfront"]==1]
```

```python
houses[houses["price"]<500000]
```

```python
houses[(houses["waterfront"]==1)&(houses["price"]<500000)]
```

```python
# find houses which have a very good view and also a very good grade
```

```python
houses[houses["view"]==4]
```

```python
houses[houses["grade"]>=11]
```

```python
houses[(houses["grade"]>=11)&(houses["view"]==4)]
```

```python
# fing high quality house with good view and sqft_living less than 3000
high_quality=houses["grade"]>=11
good_view=houses["view"]==4
smaller=houses["sqft_living"]<3500
houses[high_quality&good_view&smaller]
```

# Combining using OR Operation

```python
# homes that were built after 2014 or renovated after 2014
houses[houses["yr_built"]>=2014]
houses[houses["yr_renovated"]>=2014]
houses[(houses["yr_built"]>=2014)|(houses["yr_renovated"]>=2014)]
```

```python
# find movies directed my martin scorsese or david fincher
```

```python
netflix=pd.read_csv("../data/netflix_titles.csv",sep="|",index_col=3)
netflix.head()
```

```python
# method 1
david_fincher=netflix["director"]=="David Fincher"
martin_scorsese=netflix["director"]=="Martin Scorsese"
netflix[(david_fincher)|(martin_scorsese)]
```

```python
# method 2
netflix[netflix["director"].isin(["David Fincher","Martin Scorsese"])]
```

```python
# find movies directed my martin scorsese or david fincher with release year after 2010
```

```python
fincher=netflix["director"]=="David Fincher"
scorsese=netflix["director"]=="Martin Scorsese"
recent=netflix["release_year"]>2015
netflix[(fincher|scorsese)&(recent)]
```

# Bitwise Negation

```python
df=titanic.head()
women=df.sex=="female"
women
```

```python
~women
```

```python
df[women]
```

```python
df[~women]
```

```python
newly_built=houses["yr_built"]>=2014
newly_renovated=houses["yr_renovated"]>=2014
recent_homes=newly_built|newly_renovated
houses[recent_homes]
```

```python
houses[~recent_homes]
```

```python
# find the people who survived the titanic
```

```python
titanic[titanic.survived==0]
```

```python
titanic[~(titanic.survived==0)]
```
