# Updating Values using loc[]

```python
# we need to set the index as the name column first to use the loc
countries.loc["Brazil"]
```

```python
# specify the index column and next specify the columns to lookup using loc
# first array specifies index labels to search,second array specifies the column names
countries.loc[["Costa Rica","Panama","Honduras"],["Ladder score","Healthy life expectancy"]]
```

```python
# change the "Regional Indicator" from westrn europe to scandinavia for selected countries
```

```python
# select the column to change using loc
countries.loc[["Denmark","Sweden","Norway"],["Regional indicator"]]
```

```python
# update the column
countries.loc[["Denmark","Sweden","Norway"],["Regional indicator"]]="Scandinavia"
```

```python
countries.loc[["Denmark","Sweden","Norway"],["Regional indicator"]]
```

```python
# error possibility when using loc if not careful
countries.loc["Finland"]="LOL"
countries.loc[["Finland"]]
```

```python
# find the houses with grade as 12 and houses with view as 4
high_quality=houses["grade"]>12
great_views=houses["view"]==4
houses[high_quality&great_views]
```

```python
# add a column with name luxury level and keep as normal for all
houses['luxury_level']="normal"
```

```python
# modify the column(luxury level) and keep as ultra for the the houses with grade as 12 and houses with view as 4
```

```python
houses.loc[high_quality&great_views,['luxury_level']]='ultra'
```
