```python
import pandas as pd
titanic=pd.read_csv("../data/titanic.csv")
titanic
```

```python
titanic.pclass.value_counts()
```

```python
# data in plot is sorted based on default descending order
titanic.pclass.value_counts().plot(kind="bar")
```

```python
titanic.pclass.value_counts().sort_values().plot(kind="bar")
```

```python
# sort based on the index - the label
titanic.pclass.value_counts().sort_index()
```

```python
# sort based on the index - the label and plot
titanic.pclass.value_counts().sort_index().plot(kind="bar")
```

# loc

```python
# access data by row rather than column
```

```python
import pandas as pd
countries=pd.read_csv("../data/world-happiness-report-2021.csv")
countries.set_index("Country name",inplace=True)
countries
```

```python
# we can get a single row in dataframe
countries.loc[["Yemen"]]
```

```python
# search for multiple rows
countries.loc[["Canada","Mexico","United States"]]
```

```python
# we can get a slice of list items based on the index using the ":" notation
titanic.loc[5:7] 
```

```python
# we can combine the slice and step to display every other value between 5 and 10
titanic.loc[5:10:2]
```

```python
countries.sort_index(inplace=True)
countries
```

```python
# we can also slice text based index values using loc
# note we need to apply sort before text based slice
countries.loc["Denmark":"Egypt"]
```
