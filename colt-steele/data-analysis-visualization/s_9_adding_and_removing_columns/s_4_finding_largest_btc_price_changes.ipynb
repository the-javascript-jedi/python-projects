```python
import pandas as pd
houses=pd.read_csv("../data/kc_house_data.csv")
titanic=pd.read_csv("../data/titanic.csv")
netflix=pd.read_csv("../data/netflix_titles.csv",sep="|",index_col=3)
btc=pd.read_csv("../data/coin_Bitcoin.csv")
countries=pd.read_csv("../data/world-happiness-report-2021.csv")
```

```python
btc.set_index("Date",inplace=True)
```

```python
# find the difference between open and close prices and display in a column
btc["Change"]=btc["Close"] - btc["Open"]
```

```python
btc
```

```python
btc.sort_values("Change",ascending=False)
```

```python
# find the difference between high and low and display in a column
btc["delta"]=btc["High"]-btc["Low"]
```

```python
btc
```

```python
# find the top delta values
btc.sort_values("delta",ascending=False)
```

```python
# find top 10 changes and what day it occured
btc.sort_values("delta",ascending=False).head(10)["delta"].plot(kind="bar")
```
