```python
# # Method 1 - drop a particular column
# axis=1 or axis='columns' means the column is dropped
btc.drop(labels="Symbol",axis=1)
```

```python
# original df is not altered, use inplace=True to make changes permanent
# btc.drop(columns=["SNo","Name","Symbol"],inplace=True)
```

```python
# drop 3 columns
btc.drop(labels=["SNo","Name","Symbol"],axis="columns")
```
