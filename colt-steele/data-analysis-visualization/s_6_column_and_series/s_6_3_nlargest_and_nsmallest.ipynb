```python
# specify the largest values in the column
# default value for nlargest is 5 
houses.price.nlargest()
```

```python
# we can pass a custom row number to nlargest()
houses.price.nlargest(9)
```

```python
houses.price.nsmallest(10)
```

```python
# we can also use nlargest and nsmallest on entire datframes
# we need to send the column names in a list
houses.nlargest(10,["price"])
# the o/p will have the entire dataframe sorted by the passed columns(price)
```

```python
# find the largest bedrooms
houses.nlargest(10,["bedrooms"])
```

```python
# find the largest bedrooms and sub sort based on bathrooms
houses.nlargest(10,["bedrooms","bathrooms"])
```
